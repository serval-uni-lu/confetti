```python
from aeon import datasets
import numpy as np
from Notebooks.classifiers import fcn
```

### Load Dataset

For this example, we will employ the Ariculary Word Recognition dataset from the UCR Archive. The training dataset contains 275 cases and the test set contins 300 cases. This is a multivariate dataset with 9 different dimensions, each dimension has a time serie of 144 timesteps. The target variable has 25 different classes.

```python
train_features, train_labels = datasets.load_classification("ArticularyWordRecognition",split="train")[0], datasets.load_classification("ArticularyWordRecognition",split="train")[1]
test_features, test_labels = datasets.load_classification("ArticularyWordRecognition",split="test")[0], datasets.load_classification("ArticularyWordRecognition",split="test")[1]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Assuming y_train is your original class labels with shape (num_samples, 1)
# You may need to reshape y_train if it's in 1D array shape
y_train_reshaped = train_labels.reshape(-1, 1)

# Initialize OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform on training labels
y_train_encoded = encoder.fit_transform(y_train_reshaped).toarray()


y_test_reshaped = test_labels.reshape(-1, 1)
y_test_encoded = encoder.transform(y_test_reshaped).toarray()
```

```python
nb_classes = len(np.unique(train_labels))
input_shape = (train_features.shape[1], train_features.shape[2])
```

```python
model = fcn.Classifier_FCN(
    output_directory="/Users/alanparedescetina/Desktop/Tesis/Models/",
    input_shape=input_shape,
    nb_classes=nb_classes
    )
```

```python
model.fit(train_features,y_train_encoded,test_features,y_test_encoded)
```
